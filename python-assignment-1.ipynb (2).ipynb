````python
import numpy as np
import csv
 
# Read the CSV file

file = open("Happiness_Data.csv", "r", encoding="utf-8-sig")
reader = csv.reader(file)
rows = list(reader)
file.close()

# First row is the header
header = rows[0]
data = rows[1:]

# Find column positions
person_i = header.index("PersonID")
age_i = header.index("Age_years")
height_i = header.index("Height_cm")
weight_i = header.index("Weight_Kg")
province_i = header.index("Province")
edu_i = header.index("Education_Level")
lang_i = header.index("NumberOfLanguages")
volley_i = header.index("LikeVollyball")
table_i = header.index("LikeTableTennis")
movie_i = header.index("FavouriteMovieName")
happy_i = header.index("HappinessLevel")

# Empty lists for cleaned columns
PersonID = []
Age = []
Height = []
Weight = []
Province = []
Education_Level = []
NumberOfLanguages = []
LikeVollyball = []
LikeTableTennis = []
FavouriteMovieName = []
HappinessLevel = []
 
# Clean each row

for row in data:

    #  PersonID 
    x = row[person_i].strip().upper()

    if x.isdigit():
        n = int(x)
        if 1 <= n <= 20:
            PersonID.append("P" + str(n))
        else:
            PersonID.append(None)

    elif len(x) > 1 and x[0] == "P" and x[1:].isdigit():
        n = int(x[1:])
        if 1 <= n <= 20:
            PersonID.append("P" + str(n))
        else:
            PersonID.append(None)

    else:
        PersonID.append(None)

    #  Age 
    x = row[age_i].strip()
    if x == "" or x.lower() == "none":
        Age.append(np.nan)
    else:
        try:
            v = float(x)
            if v.is_integer() and 15 <= v <= 100:
                Age.append(int(v))
            else:
                Age.append(np.nan)
        except:
            Age.append(np.nan)

    #  Height 
    x = row[height_i].strip()
    if x == "" or x.lower() == "none":
        Height.append(np.nan)
    else:
        try:
            v = float(x)
            if 120 <= v <= 250:
                Height.append(v)
            else:
                Height.append(np.nan)
        except:
            Height.append(np.nan)

    #  Weight 
    x = row[weight_i].strip()
    if x == "" or x.lower() == "none":
        Weight.append(np.nan)
    else:
        try:
            v = float(x)
            if 30 <= v <= 250:
                Weight.append(v)
            else:
                Weight.append(np.nan)
        except:
            Weight.append(np.nan)

    #  Province 
    x = row[province_i].strip()
    if x == "Sichuang":
        x = "Sichuan"
    if x == "Shan'xi":
        x = "Shaanxi"
    Province.append(x)

    #  Education_Level 
    x = row[edu_i].strip()
    if x == "Postgraduate/ Master's Degree (PG)":
        x = "Postgraduate/ Master’s Degree (PG)"
    if x == "None":
        x = None
    Education_Level.append(x)

    #  NumberOfLanguages 
    x = row[lang_i].strip()
    if x == "" or x.lower() == "none":
        NumberOfLanguages.append(np.nan)
    else:
        try:
            v = float(x)
            if v.is_integer() and 1 <= v <= 20:
                NumberOfLanguages.append(int(v))
            else:
                NumberOfLanguages.append(np.nan)
        except:
            NumberOfLanguages.append(np.nan)

    #  LikeVollyball 
    x = row[volley_i].strip().upper()
    if x == "TRUE":
        LikeVollyball.append(True)
    elif x == "FALSE":
        LikeVollyball.append(False)
    else:
        LikeVollyball.append(None)

    #  LikeTableTennis 
    x = row[table_i].strip().upper()
    if x == "TRUE":
        LikeTableTennis.append(True)
    elif x == "FALSE":
        LikeTableTennis.append(False)
    else:
        LikeTableTennis.append(None)

    #  FavouriteMovieName 
    x = row[movie_i].strip()
    if x == "" or x.lower() == "none":
        FavouriteMovieName.append(None)
    else:
        FavouriteMovieName.append(x)

    #  HappinessLevel 
    x = row[happy_i].strip()
    if x == "" or x.lower() == "none":
        HappinessLevel.append(np.nan)
    else:
        try:
            v = float(x)
            if v.is_integer() and 1 <= v <= 10:
                HappinessLevel.append(int(v))
            else:
                HappinessLevel.append(np.nan)
        except:
            HappinessLevel.append(np.nan)

# Convert lists to numpy arrays
PersonID = np.array(PersonID, dtype=object)
Age = np.array(Age, dtype=float)
Height = np.array(Height, dtype=float)
Weight = np.array(Weight, dtype=float)
Province = np.array(Province, dtype=object)
Education_Level = np.array(Education_Level, dtype=object)
NumberOfLanguages = np.array(NumberOfLanguages, dtype=float)
LikeVollyball = np.array(LikeVollyball, dtype=object)
LikeTableTennis = np.array(LikeTableTennis, dtype=object)
FavouriteMovieName = np.array(FavouriteMovieName, dtype=object)
HappinessLevel = np.array(HappinessLevel, dtype=float)

# 
# Print results
# 
print("Unique PersonID values:")
print(np.unique(PersonID.astype(str)))

print("\nUnusual Age values changed to nan:")
print(Age[np.isnan(Age)])

print("\nUnusual Height values changed to nan:")
print(Height[np.isnan(Height)])

print("\nUnusual Weight values changed to nan:")
print(Weight[np.isnan(Weight)])

print("\nUnusual NumberOfLanguages values changed to nan:")
print(NumberOfLanguages[np.isnan(NumberOfLanguages)])

print("\nUnique values in LikeVollyball:")
print(np.unique(LikeVollyball.astype(str)))

print("\nUnique values in LikeTableTennis:")
print(np.unique(LikeTableTennis.astype(str)))

print("\nUnique Province values:")
print(np.unique(Province))

print("\nUnique Education_Level values:")
print(np.unique(Education_Level.astype(str)))

print("\nUnique HappinessLevel values:")
print(np.unique(HappinessLevel[~np.isnan(HappinessLevel)]))
````

````python
import numpy as np

def sum_with_none(x):
    sum_x = 0
    n_x = 0
    
    for val in x:
        if val is None:
            continue
        if isinstance(val, float) and np.isnan(val):
            continue
        
        sum_x += val
        n_x += 1
    
    return sum_x, n_x

def mean_with_none(x):
    sum_x, n_x = sum_with_none(x)
    
    if n_x == 0:
        return None
    
    return sum_x / n_x

def std_with_none(x):
    import numpy as np
    
    mean_x = mean_with_none(x)
    
    if mean_x is None:
        return None
    
    sum_sq_diff = 0
    n_x = 0
    
    for val in x:
        if val is None:
            continue
        if isinstance(val, float) and np.isnan(val):
            continue
        
        sum_sq_diff += (val - mean_x) ** 2
        n_x += 1
    
    if n_x == 0:
        return None
    
    variance = sum_sq_diff / n_x
    return np.sqrt(variance)


# Example test data
x = [1, 2, 3, None, 3, np.nan]

# Testing sum_with_none
sum_x, n_x = sum_with_none(x)
print("Sum:", sum_x)        
print("Count:", n_x)       

# Testing mean_with_none
mean_x = mean_with_none(x)
print("Mean:", mean_x)      

# Testing std_with_none
std_x = std_with_none(x)
print("Standard Deviation:", std_x)  
````

````python
import csv
import matplotlib.pyplot as plt

# Store education levels
education = []

with open('Happiness_Data.csv', 'r') as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        education.append(row['Education_Level'])

# Count each category
counts = {}

for e in education:
    if e in counts:
        counts[e] += 1
    else:
        counts[e] = 1

# Prepare data for pie chart
labels = list(counts.keys())
sizes = list(counts.values())

# Explode Post Graduate
explode = []
for label in labels:
    if label == 'Post Graduate':
        explode.append(0.2)   # highlight this slice
    else:
        explode.append(0)

# Plot pie chart
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%')
plt.title("Education Level Distribution")
plt.show()
````

````python
import csv
import numpy as np

# Step 1: Read file
rows = []
with open('Happiness_Data.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        rows.append(row)

# Step 2: Convert to numpy array
data = np.array(rows[1:])  # skip header

# Step 3: Extract columns
education = data[:, 8]
movies = data[:, 12]

# Step 4: Filter PG and above
mask = np.array([
    ("PG" in e) or ("Postgraduate" in e) or ("Master" in e) or ("PhD" in e) or ("Doctorate" in e)
    for e in education
])

filtered_movies = movies[mask]

# Step 5: Clean data
filtered_movies = np.array([
    m for m in filtered_movies if m not in ["None", "", "nan"]
])

# Step 6: Count frequency
values, counts = np.unique(filtered_movies, return_counts=True)

# Step 7: Find most popular
most_popular_movie = values[np.argmax(counts)]
most_count = np.max(counts)

# Step 8: Output
print("Most popular movie among PG and above:")
print(most_popular_movie)
print("Count:", most_count)
````

````python
#5.3A

import csv
import numpy as np
import matplotlib.pyplot as plt


# read csv file

file_path = "Happiness_Data.csv"

with open(file_path, "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    province_list = []
    happiness_list = []

    for row in reader:
        province = row["Province"].strip()
        happiness = row["HappinessLevel"].strip()

        # store province
        if province == "" or province.lower() == "none" or province.lower() == "nan":
            province_list.append("None")
        else:
            province_list.append(province)

        # store happiness
        try:
            happiness_list.append(float(happiness))
        except:
            happiness_list.append(np.nan)

# convert to numpy arrays
province_arr = np.array(province_list)
happiness_arr = np.array(happiness_list)

unique_provinces, counts = np.unique(province_arr, return_counts=True)

plt.figure(figsize=(12, 6))
plt.bar(unique_provinces, counts)
plt.xlabel("Province")
plt.ylabel("Number of People")
plt.title("Number of People in Each Province")
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()

print("Number of people in each province:")
for i in range(len(unique_provinces)):
    print(unique_provinces[i], ":", counts[i])

# 
# Q5.3B - Happiness Index of each province

# remove None province
mask = province_arr != "None"
province_valid = province_arr[mask]
happiness_valid = happiness_arr[mask]

unique_valid_provinces = np.unique(province_valid)

avg_happiness = []

for prov in unique_valid_provinces:
    scores = happiness_valid[province_valid == prov]
    scores = scores[~np.isnan(scores)]   # remove nan happiness values

    if len(scores) > 0:
        avg_happiness.append(np.mean(scores))
    else:
        avg_happiness.append(np.nan)

avg_happiness = np.array(avg_happiness)

# remove provinces whose average is nan
valid_avg_mask = ~np.isnan(avg_happiness)
unique_valid_provinces = unique_valid_provinces[valid_avg_mask]
avg_happiness = avg_happiness[valid_avg_mask]

# sort provinces by average happiness
sorted_index = np.argsort(avg_happiness)

bottom5_index = sorted_index[:5]
top5_index = sorted_index[-5:]

bottom5_provinces = unique_valid_provinces[bottom5_index]
bottom5_scores = avg_happiness[bottom5_index]

top5_provinces = unique_valid_provinces[top5_index]
top5_scores = avg_happiness[top5_index]

# plot bottom 5
plt.figure(figsize=(8, 5))
plt.bar(bottom5_provinces, bottom5_scores)
plt.xlabel("Province")
plt.ylabel("Average Happiness Score")
plt.title("Bottom 5 Provinces by Happiness Index")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# plot top 5
plt.figure(figsize=(8, 5))
plt.bar(top5_provinces, top5_scores)
plt.xlabel("Province")
plt.ylabel("Average Happiness Score")
plt.title("Top 5 Provinces by Happiness Index")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\nBottom 5 provinces by happiness index:")
for i in range(len(bottom5_provinces)):
    print(bottom5_provinces[i], ":", round(bottom5_scores[i], 2))

print("\nTop 5 provinces by happiness index:")
for i in range(len(top5_provinces)):
    print(top5_provinces[i], ":", round(top5_scores[i], 2))
````

````python
import csv
import numpy as np
import matplotlib.pyplot as plt

file_path = "Happiness_Data.csv"

#  Education Level Mapping
# Assign numerical values (higher = more educated)
education_map = {
    "high school": 1,
    "undergraduate": 2,
    "bachelor": 2,
    "post graduate": 3,
    "master": 3,
    "phd": 4,
    "doctorate": 4
}

#  Load Data 
with open(file_path, newline='', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    rows = list(reader)

#  Store education scores per province 
province_education = {}

for row in rows:
    province = row["Province"].strip()
    edu = row["Education_Level"].strip().lower()
    
    # Ignore invalid values
    if province == "" or province.lower() in ["none", "nan"]:
        continue
    
    if edu not in education_map:
        continue
    
    score = education_map[edu]
    
    if province not in province_education:
        province_education[province] = []
    
    province_education[province].append(score)

#  Compute average education per province 
avg_education = {}

for province, scores in province_education.items():
    avg_education[province] = np.mean(scores)

#  Get Top 5 Provinces 
sorted_provinces = sorted(avg_education.items(), key=lambda x: x[1], reverse=True)
top_5 = sorted_provinces[:5]

labels = [item[0] for item in top_5]
values = [item[1] for item in top_5]

#  Plot 
plt.figure()
plt.bar(labels, values)
plt.xticks(rotation=45)

plt.title("Top 5 Provinces by Education Level")
plt.xlabel("Province")
plt.ylabel("Average Education Score")

plt.show()
````

````python
import csv
import numpy as np

#  Load data 
with open("Happiness_Data.csv", "r", encoding="utf-8-sig") as f:
    data = list(csv.DictReader(f))

#  Simple converters 
def to_float(x):
    if x == "" or x == "None":
        return np.nan
    return float(x)

def to_bool(x):
    if x == "TRUE":
        return 1.0
    elif x == "FALSE":
        return 0.0
    return np.nan

edu_map = {
    "No School": 0,
    "High School": 1,
    "Undergraduate Degree (UG)": 2,
    "Postgraduate/ Master's Degree (PG)": 3
}

def to_edu(x):
    return edu_map.get(x, np.nan)

#  Closeness function 
def closeness(p1, p2):
    """
    Compute similarity between two people using Euclidean distance.

    Steps:
    - Select attributes
    - Convert to numbers
    - Handle missing values (mean)
    - Scale (z-score)
    - Compute distance
    """

    # select attributes as simple lists
    a = [
        to_float(p1["Age_years"]),
        to_float(p1["Height_cm"]),
        to_float(p1["Weight_Kg"]),
        to_float(p1["NumberOfLanguages"]),
        to_float(p1["HappinessLevel"]),
        to_bool(p1["LikeVollyball"]),
        to_bool(p1["LikeTableTennis"]),
        to_edu(p1["Education_Level"])
    ]

    b = [
        to_float(p2["Age_years"]),
        to_float(p2["Height_cm"]),
        to_float(p2["Weight_Kg"]),
        to_float(p2["NumberOfLanguages"]),
        to_float(p2["HappinessLevel"]),
        to_bool(p2["LikeVollyball"]),
        to_bool(p2["LikeTableTennis"]),
        to_edu(p2["Education_Level"])
    ]

    a = np.array(a, dtype=float)
    b = np.array(b, dtype=float)

    # handle missing values
    mean = np.nanmean([a, b], axis=0)
    a = np.where(np.isnan(a), mean, a)
    b = np.where(np.isnan(b), mean, b)

    # scaling
    std = np.std([a, b], axis=0)
    std[std == 0] = 1

    a = (a - mean) / std
    b = (b - mean) / std

    #  distance
    return np.sqrt(np.sum((a - b) ** 2))


#  Test (PRINT RESULT) 
p1 = data[0]
p2 = data[1]

distance = closeness(p1, p2)

print("Person 1:", p1)
print("Person 2:", p2)
print("Distance:", distance)
````
