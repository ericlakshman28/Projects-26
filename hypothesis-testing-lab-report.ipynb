````python
print('8) Research  Question: Does semanticity influence attention scores, and does this effect vary with noise level')

print('Do people perform differently when sentences are semantic or non-semantic and does noise change this difference')

print('Why I chose this as my research: Semanticity is a crucial experimental variable that may influence the ease with which participants process auditory information. It is anticipated that meaningful (semantic) stimuli may enhance attention, especially in difficult listening environments. Examining its interaction with noise facilitates a more profound comprehension of cognitive processing across diverse levels of difficulty')

print('Hypothesis:')
print('Null Hypothesis: Semanticity does not affect attention, and its effect does not depend on noise')
print('Alternate Hypothesis: Semanticity affects attention, and this effect may depend on noise level')

print('Method: I will calculate the mean attention scores for each combination of semanticity and noise level. I will use a bar chart to visualise the relationship between these variables. This will give a picture of any patterns and interaction effects between semanitcity and noise')
print(' I will also use independent t-tests comparing semantic and non-semantic conditions at each noise level, this wll allows us to see if attention and this effect changes under noise conditions')
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("PhyAAt_AttentionScoreData_v1.csv")

# Calculate means
means = df.groupby(["Semanticity", "SNRdB"])["AttentionScore"].mean().unstack()

print("Mean attention scores:\n", means)

# Plot
means.plot(kind="bar")
plt.title("Interaction between Semanticity and Noise on Attention")
plt.xlabel("Semanticity (0 = non-semantic, 1 = semantic)")
plt.ylabel("Mean Attention Score")
plt.show()

#Interpretation of Graphs:

print('The bar chart shows how attention scores will vary across semantic and non semantic stimuli. In general, attention scores increase as noise decreases which confirms that listening conditions influence performance. Differences between semantic and non-semantic can also be observed. Semantic conetnt may influence attention') 

#Statiscal Test:

print("\nStatistical Tests (semantic vs non-semantic at each noise level):")

for snr in df ["SNRdB"].unique():
    group0 = df[(df["Semanticity"] == 0) & (df["SNRdB"] == snr)] ["AttentionScore"]
    group1 = df[(df["Semanticity"] == 1) & (df["SNRdB"] == snr)] ["AttentionScore"]

    t,p == ttest_ind(group0, group1)

    print(f"SNR = {snr}: t = {round(t,2)}, p = {p}")

#Interpretation:

print('since the p value is 0.876 which is greater than 0.05 this means no significant difference between semantic vs non-semantic, at any noise level and no interaction')


#Conclusion:

print('Conclusion: The analysis from the graphs show that attention scores as increased as noise decreased for both semantic and non-smenatic stimuli. However the statistical test revealed no difference between semantic and non-semantic conditions at any noise level. This indicates that semanticity does not have a big effect on attention performance in regards to this dataset. Finally, since rhe results were consistent across all noise levels, there is no evidence of an interaction between semanticity and noise. Noise seems to be the key factor in influence of attention rather than semantic content')
````
